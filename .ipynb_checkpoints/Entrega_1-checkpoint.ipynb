```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np

# Lendo o arquivo de dados
dados = pd.read_excel('Dados de estudantes da UFV - 2022-12-12.ods')
```

```python
# Visualizando as primeiras linhas do DataFrame
print( dados.head())

# Verificando valores únicos e contagem para cada atributo
for coluna in dados.columns:
    print("\n"*3 + "{}\n Variedade: {}\n Dados: {}".format(coluna, dados[coluna].nunique(), dados[coluna].unique()))
```
