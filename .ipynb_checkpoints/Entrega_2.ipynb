#### Tipo de dado
    - Transformando o tipo de dado para string

```python
df['Municipio_Nascimento'] = df['Municipio_Nascimento'].astype(str)
```

#### Nome dos municipios
    - Tratamento de nomes de municipios

***Removendo os espaços em branco no início e no fim de cada nome de municipio***

```python
df['Municipio_Nascimento'] = df['Municipio_Nascimento'].str.strip()
```

***Removendo os espaços em branco duplicados no meio de cada nome de municipio***

```python
df['Municipio_Nascimento'] = df['Municipio_Nascimento'].str.replace(' +', ' ', regex=True)
```

***Municipios que estão em uppercase serão colocados em lower-case, com exceção da suas iniciais***

```python
if df['Municipio_Nascimento'].str.isupper().any():
    df['Municipio_Nascimento'] = df['Municipio_Nascimento'].str.lower()
    df['Municipio_Nascimento'] = df['Municipio_Nascimento'].str.title()
```

***Colocando todas as cidades em uma lista***

```python
munUFV = df['Municipio_Nascimento'].unique().tolist()
len(munUFV)
```

***Lista de municípios do IBGE - Tratamento***

```python
url = 'https://servicodados.ibge.gov.br/api/v1/localidades/municipios'
r = requests.get(url)
munExistente = json.loads(r.text)
munExistente = [municipio['nome'] for municipio in munExistente]
```

***Verificando se os municípios dos dados de alunos existem na lista do IBGE***

```python
munNaoExistente = []
for municipio in munUFV:
    if municipio not in munExistente:
        munNaoExistente.append(municipio)
print(f'Quantidade de municípios que não estão na lista do IBGE: {len(munNaoExistente)}')
```

***Tratamento dos possíveis municipios que não existem. Similaridades acima de 90 entre os municipios que não existem e os municipios do IBGE***
***usando a biblioteca fuzzywuzzy, serão substituídos***

```python
munSim = []

for municipio in munNaoExistente:
    for municipioIBGE in munExistente:
        if fuzz.ratio(municipio, municipioIBGE) > 90:
            munSim.append((municipio, municipioIBGE))
print(f'Quantidade de municípios que não estão na lista do IBGE: {len(munSim)}')
```

***Usando similaridades para substituir os municipios que não existem para os municipios do IBGE***

```python
for municipio in munSim:
    df['Municipio_Nascimento'] = df['Municipio_Nascimento'].str.replace(municipio[0], municipio[1])
```

##### Excluindo os municipios trocados, da lista que não existem, pois agora, eles estão no df

```python
for municipio in munSim:
    if municipio[0] in munNaoExistente:
        munNaoExistente.remove(municipio[0])
print(f'Quantidade de municípios que ainda não estão na lista do IBGE: {len(munNaoExistente)}')
```

***O restante dos municipios que não existem serão excluídos, pois não possuem similaridades com os municipios do IBGE ou são de outros países***

```python
for municipio in munNaoExistente:
    df.drop(df[df['Municipio_Nascimento'] == municipio].index, inplace=True)
```

### Respondendo algumas perguntas

#### 2 Existem diferenças significativas nas taxas de evasão entre diferentes gêneros? E de grupos raciais?

```python
dfevasao = df[df['Situacao_Aluno_Agrupada'] == 'Evasão']
evasaoM = dfevasao[dfevasao['Sexo'] == 'M']
evasaoF = dfevasao[dfevasao['Sexo'] == 'F']

porcentagemM = (len(evasaoM) / len(df[df['Sexo'] == 'M'])) * 100
porcentagemF = (len(evasaoF) / len(df[df['Sexo'] == 'F'])) * 100

print(f"Porcentagem de evasão de homens: {porcentagemM:.2f}%")
print(f"Porcentagem de evasão de mulheres: {porcentagemF:.2f}%")
```

##### Em geral, os homens evadem 10% a mais que as mulheres

#### 1 Qual é a taxa de evasão geral em cada curso? Compare. (essa comparação pode ser agrupada por campus ou área (exatas, humanas, etc))

```python
# Todos os cursos, ignorando Licenciatura

dfTotalalunos = df.groupby('Curso').size().reset_index().rename(columns={0: 'Total_Alunos'})
dfEvasaoCurso = df[df['Situacao_Aluno_Agrupada'] == 'Evasão'].groupby('Curso').size().reset_index().rename(columns={0: 'Evasao'})
dfEvasaoCurso = dfEvasaoCurso.merge(dfTotalalunos, on='Curso')
dfEvasaoCurso['Taxa_Evasao'] = (dfEvasaoCurso['Evasao'] / dfEvasaoCurso['Total_Alunos']) * 100
dfEvasaoCurso = dfEvasaoCurso.sort_values(by='Taxa_Evasao', ascending=False)
dfEvasaoCurso = dfEvasaoCurso[dfEvasaoCurso['Curso'].str.contains('Licenciatura') == False]
dfEvasaoCurso
```

```python
# Taxa de evasão por Campus - Exceção de Licenciaturas

dfTotalCAV = df[df['Campus'] == 'CAV'].groupby('Curso').size().reset_index().rename(columns={0: 'Total_Alunos'})
dfTotalCAF = df[df['Campus'] == 'CAF'].groupby('Curso').size().reset_index().rename(columns={0: 'Total_Alunos'})
dfTotalCRP = df[df['Campus'] == 'CRP'].groupby('Curso').size().reset_index().rename(columns={0: 'Total_Alunos'})

dfTotalAlunosCAV = df[df['Campus'] == 'CAV']
dfTotalAlunosCAV = dfTotalAlunosCAV[~dfTotalAlunosCAV['Curso'].str.contains('Licenciatura', case=False)]
dfTotalAlunosCAV = dfTotalAlunosCAV.groupby('Curso').size().sum()
dfEvasaoTotalCAV = df[(df['Campus'] == 'CAV') & (df['Situacao_Aluno_Agrupada'] == 'Evasão')]
dfEvasaoTotalCAV = dfEvasaoTotalCAV[~dfEvasaoTotalCAV['Curso'].str.contains('Licenciatura', case=False)]
dfEvasaoTotalCAV = dfEvasaoTotalCAV.groupby('Curso').size().sum()
dfEvasaoCAV = dfEvasaoTotalCAV / dfTotalAlunosCAV * 100
dfEvasaoCAV = pd.DataFrame({'Campus': ['CAV'], 'Taxa_Evasao': [dfEvasaoCAV]})

dfTotalAlunosCAF = df[df['Campus'] == 'CAF']
dfTotalAlunosCAF = dfTotalAlunosCAF[~dfTotalAlunosCAF['Curso'].str.contains('Licenciatura', case=False)]
dfTotalAlunosCAF = dfTotalAlunosCAF.groupby('Curso').size().sum()
dfEvasaoTotalCAF = df[(df['Campus'] == 'CAF') & (df['Situacao_Aluno_Agrupada'] == 'Evasão')]
dfEvasaoTotalCAF = dfEvasaoTotalCAF[~dfEvasaoTotalCAF['Curso'].str.contains('Licenciatura', case=False)]
dfEvasaoTotalCAF = dfEvasaoTotalCAF.groupby('Curso').size().sum()
dfEvasaoCAF = dfEvasaoTotalCAF / dfTotalAlunosCAF * 100
dfEvasaoCAF = pd.DataFrame({'Campus': ['CAF'], 'Taxa_Evasao': [dfEvasaoCAF]})


dfTotalAlunosCRP = df[df['Campus'] == 'CRP']
dfTotalAlunosCRP = dfTotalAlunosCRP[~dfTotalAlunosCRP['Curso'].str.contains('Licenciatura', case=False)]
dfTotalAlunosCRP = dfTotalAlunosCRP.groupby('Curso').size().sum()
dfEvasaoTotalCRP = df[(df['Campus'] == 'CRP') & (df['Situacao_Aluno_Agrupada'] == 'Evasão')]
dfEvasaoTotalCRP = dfEvasaoTotalCRP[~dfEvasaoTotalCRP['Curso'].str.contains('Licenciatura', case=False)]
dfEvasaoTotalCRP = dfEvasaoTotalCRP.groupby('Curso').size().sum()
dfEvasaoCRP = dfEvasaoTotalCRP / dfTotalAlunosCRP * 100
dfEvasaoCRP = pd.DataFrame({'Campus': ['CRP'], 'Taxa_Evasao': [dfEvasaoCRP]})

# Concatenando os dataframes
dfEvasaoCampus = pd.concat([dfEvasaoCAV, dfEvasaoCAF, dfEvasaoCRP])
dfEvasaoCampus = dfEvasaoCampus.sort_values(by='Taxa_Evasao', ascending=False)
dfEvasaoCampus
```

#### A taxa de Evasão é maior no campus CAF e CRP, e menor no CAV
